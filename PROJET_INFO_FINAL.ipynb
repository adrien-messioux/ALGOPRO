```python
def min(A):
    m = A[0]
    indice_min = 0
    for i in range (len(A)):
        if A[i] <= m:
            m = A[i]
            indice_min = i
    return m,int(indice_min)
```

```python
def max(A):
    M = A[0]
    indice_max = 0
    for i in range (len(A)):
        if A[i] >= M:
            M = A[i]
            indice_max = i
    return M,int(indice_max)
```

```python
def moyenne_arithmetique (A):
    somme = 0
    for i in range (len(A)):
        somme += A[i]
    return somme / len(A)
```

```python
def moyenne_geometrique (A):
    somme = 0
    for i in range (len(A)):
        somme += np.log (A[i])
    return np.exp (somme/len(A))

moyenne(L2)
```

```python
def variance(A):
    v=0
    nb = len(A)
    for k in range(nb):
        v += (A[k]-moyenne(A))**2
    return v/nb

variance(noise[0])
```

```python
def ecart_type(A):
    sigma = np.sqrt(variance(A))
    return sigma

ecart_type(noise[0])
```

```python
def mediane(A):
    n=len(A)
    if n%2==0:
        a=A[n//2]
        b=A[(n//2)+1]
        mediane=moyenne((a,b))
    else:
        mediane=A[n//2]
    return mediane

mediane(noise[1])
```

```python
# fonction qui prend les valeurs de N et les divise pour 3 jours, soit en 5 sous-liste

nb_divise = 5
instant_separation = []
instant_max = max(temps)[0]
pas = instant_max / nb_divise
N_divise = [ [ 0 for j in range (nb_divise) ] for _ in range (6) ]

for j in range (nb_divise+1):
    instant_separation.append( pas*j )
    

    
def est_entre (instant):
    for j in range (nb_divise):
        if (instant_separation[j] <= instant) and (instant_separation[j+1] > instant):
            return j
        elif instant == instant_max :
            return j

for i in range (len(temps)):
    j = est_entre(temps[i])
    k = L1[i] - 1
    N_divise[k][j] += 1
            
N_divise
```
